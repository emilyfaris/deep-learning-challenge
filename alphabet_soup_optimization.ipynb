## Preprocessing

```python
# Importing dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import pandas as pd
import tensorflow as tf

# Setting random seed for replicable results
tf.random.set_seed(42)

#  Importing and reading the charity_data.csv
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Dropping non-beneficial ID columns
application_df = application_df.drop(columns=['EIN', 'NAME'])
application_df.head()
```

```python
# Determining number of unique values in each column
application_df.nunique()
```

```python
# Looking at APPLICATION_TYPE value counts to identify and replace with "Other"
app_value_counts = application_df['APPLICATION_TYPE'].value_counts()
app_value_counts
```

```python
# Choosing a cutoff value and creating a list of application types to be replaced with "Other"
application_types_to_replace = app_value_counts[app_value_counts <= 156].index.tolist()

# Replacing in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Checking replacement
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Looking at CLASSIFICATION value counts to identify and replace with "Other"
class_value_counts = application_df['CLASSIFICATION'].value_counts()
class_value_counts
```

```python
# Looking at CLASSIFICATION value counts >1
class_value_counts[class_value_counts > 1]
```

```python
# Choosing a cutoff value and creating a list of classifications to be replaced
classifications_to_replace = class_value_counts[class_value_counts <= 777].index.tolist()

# Replacing in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Checking to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df.head()
```

```python
# Converting categorical data to numeric
dummies_df = pd.get_dummies(application_df, dtype='float')

dummies_df.head()
```

```python
# Splitting preprocessed data into features and target arrays
y = dummies_df['IS_SUCCESSFUL']
X = dummies_df.drop(columns=['IS_SUCCESSFUL'])

# Splitting preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# Creating StandardScaler instance
scaler = StandardScaler()

# Fitting StandardScaler
X_scaler = scaler.fit(X_train)

# Scaling the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Original Model

```python
## Keeping original model outputs for reference

# Defining original model
number_input_features = X_train.shape[1]
input_nodes = 80
hidden_layer_nodes = 30

nn = tf.keras.models.Sequential()

# Input layer
nn.add(
    tf.keras.layers.Dense(units=input_nodes, input_dim=number_input_features, activation='relu')
)

# Hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_layer_nodes, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Checking the structure of the model
nn.summary()
```

```python
# Compiling the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Creating callback to save model's weight every 5 epochs
checkpoints = tf.keras.callbacks.ModelCheckpoint('AlphabetSoupCharity.h5', save_weights_only=False, save_freq=5)
# Training model
fit_model = nn.fit(X_train, y_train, epochs=100, callbacks=checkpoints)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Attempt 3: Using keras-tuner to find best hyperparameters to improve accuracy
# Create a method that creates a new Sequential model with hyperparameter options
def create_model(hp):
    nn_model = tf.keras.models.Sequential()

    # Allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','tanh','sigmoid'])

    # Allow kerastuner to decide number of neurons in first layer
    nn_model.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=1,
        max_value=10,
        step=2), activation=activation, input_dim=number_input_features))

    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 6)):
        nn_model.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=1,
            max_value=10,
            step=2),
            activation=activation))

    nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Compile the model
    nn_model.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])

    return nn_model
```

```python
# Import the kerastuner library
import keras_tuner as kt

tuner = kt.Hyperband(
    create_model,
    objective="val_accuracy",
    max_epochs=20,
    hyperband_iterations=2)
```

```python
# Run the kerastuner search for best hyperparameters
tuner.search(X_train_scaled,y_train,epochs=20,validation_data=(X_test_scaled,y_test))
```

```python
# Get best model hyperparameters
best_hyper = tuner.get_best_hyperparameters(1)[0]
best_hyper.values
```

```python
# Evaluate best model against full test data
best_model = tuner.get_best_models(1)[0]
model_loss, model_accuracy = best_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
